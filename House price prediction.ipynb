 Importing the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor 
```

Loading dataset

```python
house_price_dataset = sklearn.datasets.load_boston()
print(house_price_dataset)
```

Loading it into a dataframe

```python
#As we can see that, since the dataset is in a non-usable form, we are going to load this dataset into a structured dataframe.
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
house_price_dataframe.head()
```

```python
#inserting the label (Target column from dataset) into the dataframe
house_price_dataframe["Price"] = house_price_dataset.target
house_price_dataframe.head()
```

Making some Analysation

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.describe()
```

```python
#finding CORRELATION between each columns or features
correlation = house_price_dataframe.corr()

#creating a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws ={'size':10}, cmap = 'Blues' )
```

Data and Label split

```python
X =  house_price_dataframe.drop(['Price'], axis = 1)
Y = house_price_dataframe["Price"]
```

Train and test spliting

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 2)
```

Actual Training with XGBoost Regressor model

```python
model = XGBRegressor()
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#training data acuracy score with R squared and MAE error:
training_data_prediction = model.predict(X_train)

#R squared error:
score_1 = metrics.r2_score(Y_train, training_data_prediction)
#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error's score:", score_1)
print("MAE's Score:", score_2)
```

```python
#plotting the training data accuracy
plt.scatter(Y_train, training_data_prediction)
plt.xlabel = ("Actual Prices")
plt.ylabel = ("Predicted Prices")
plt.title("Actual Price VS Predicted Price")
plt.show
```

```python
#testing data acuracy score with R squared and MAE error:
testing_data_prediction = model.predict(X_test)

#R squared error:
score_3 = metrics.r2_score(Y_test, testing_data_prediction)
#Mean absolute error:
score_4 = metrics.mean_absolute_error(Y_test, testing_data_prediction)

print("R squared error's score:", score_3)
print("MAE's Score:", score_4)
```

```python
#plotting the testing data accuracy
plt.scatter(Y_test, testing_data_prediction)
plt.xlabel = ("Actual Prices")
plt.ylabel = ("Predicted Prices")
plt.title("Predicted test data accuracy")
plt.show
```

We got a good accuraccy score. And in the above graph, we can see that how close our model's predicts and the acutal values are.
